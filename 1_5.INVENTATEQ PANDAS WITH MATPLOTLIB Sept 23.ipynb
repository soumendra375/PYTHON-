```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### BASIC GRAPH

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize your Graph (dpi specifies pixels per inch. when saving probably should use 300 if possible)
plt.figure(figsize=(5,3), dpi=300)

plt.plot(x,y)

#show plot
plt.show()

#Add a title (specify font parameters with fontdict)
plt.title('Our First Graph')
#plt.title('Our First Graph!', fontdict={'fontname': 'Cambria', 'fontsize': 20})

## X and Y Labels
plt.xlabel('X Axis')
#plt.xlabel('X Axis', fontdict={'fontname': 'Cambria'})
plt.ylabel('Y Axis')
#plt.ylabel('Y Axis', fontdict={'fontname': 'Cambria'})

##HOW TO CHANGE OUR TEXT

##X, Y axis Tichmarks (scale of your graph)
plt.xticks([0,1,2,3,4,])
#plt.xticks([0,1,2,3,4, 1000])
##plt.yticks([0,2,4,6,8,10])
#plt.yticks([0,2,4,6,8,100])
plt.yticks([0,2,4,6,8,10])
#plt.plot([1,2,3],[2,4,6])

## Add a Legend
plt.legend()

##label
#x = [0,1,2,3,4]
#y = [0,2,4,6,8]
#plt.plot(x,y,label='2x')
#plt.plot(x,y,label='2x', color='red')
#plt.plot(x,y,label='2x', color='red', linewidth=2)

##Keyword argument notation
#plt.plot(x,y,label='2x', color='red', linewidth=2, marker='.', markersize=10)
#plt.plot(x,y,label='2x', color='red', linewidth=2, marker='.', markersize=10, markeredgecolor='blue')
plt.plot(x,y,label='2x', color='blue', linewidth=5, marker='.', linestyle='--', markersize=20, markeredgecolor='blue')

##use shorthand notation
#fmt = '[color][marker][living]'
#plt.plot(x,y, 'r.--', label='2x')
plt.plot(x,y, 'b^--', label='2x')

## Line number two
x2 = np.arange(0,4.5,0.5)
print(x2)
plt.plot(x2, x2**2, 'r', label='x^2')

##Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'r', label='x^2')

##Plot remainder of graph
plt.plot(x2[4:], x2[4:]**2, 'r--')

##Save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig('mygraph.png', dpi=100)
```

### BAR CHART

```python
labels = ['A', 'B', 'C']
values =[1,4,2]

bars = plt.bar(labels, values)

patterns = ['/', 'o', '*']
for bar in bars:
    bar.set_hatch(patterns.pop())

#Hatching
bars[0].set_hatch('/')
#Circle
bars[1].set_hatch('o')
#Stars
bars[2].set_hatch('*')

plt.figure(figsize=(6,4))

plt.legend

plt.show()
```
