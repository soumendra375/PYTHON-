```python
#from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
```

```python
# LINE PLOT OR LINE CHART
x = [1,2,3]
y = [4,5,2]
plt.plot(x,y)
```

```python
# INSERTING TITLE , XLABEL AND YLABEL
x=[1,2,3]
y=[4,5,2]
plt.plot(x,y)
plt.title('line graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
x1=[1,2,3]
y1=[4,5,2]
x2=[6,7,8]
y2=[9,10,7]
plt.plot(x1,y1,label='first',color='r',linewidth=5)
plt.plot(x2,y2,label='second',color='m',linewidth=5)
plt.title('style graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.grid(True,color='lightblue')
plt.show()
```

### CATEGORICAL PLOT OR BARPLOT (categorical data - help of bar)

```python
x1=[1,3,5,7]
y1=[5,2,6,6]
x2=[2,4,6,8]
y2=[6,3,5,2]
plt.bar(x1,y1,label='first')
plt.bar(x2,y2,label='second')
plt.legend()
plt.xlabel('bar numbers')
plt.ylabel('bar height')
plt.title('bar chart')
plt.show()
```

### HISTOGRAM PLOT(CONTINOUS DATA DISTRIBUTION PLOT- where data is more or less , data changes over time and also used for univariate distribution)

```python
sale_values=[12,10,5,7,0,14,15,24,10,11,15,30,18,11,10,18,30]
bins = [0,5,10,15,20,25,30]
plt.hist(sale_values,bins,histtype= 'bar', rwidth=0.6)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('histogram graph')
plt.show()
```

### SCATTER PLOT (to check the correlation between two variable/multiple data)

```python
x=[1,3,4,8,5,6,7]
y=[3,5,6,7,8,9,3]
plt.scatter(x,y,color='b')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('scatter plot')
plt.show()
```

### pie chart(showing proportion of data distribution)

```python
Mylabels=('python','c++', 'ruby', 'java')
sizes=[215,130,245,210]
Mycolors=['gold','yellowgreen','lightcoral','lightskyblue']

plt.pie(sizes,labels=Mylabels,colors=Mycolors,autopct='%1.2f%%') #autopct- autopercentile
plt.show()
```
